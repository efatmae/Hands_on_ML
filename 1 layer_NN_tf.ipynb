```python
#low level TF API
import tensorflow as tf

#construction phase
n_inputs = 28 * 28 #MNIST features
n_hidden1 = 300
n_hidden2 = 100
n_outputs = 10
```

```python
#building placeholder nodes to represent training and targets

X = tf.placeholder(tf.float32, shape=(None, n_inputs), name= "X") #None is the no. instances not known yet
y = tf.placeholder(tf.int64, shape=(None), name= 'y')

#construction the network
#creating layers
def neuron_layer (X, n_neurons, name, activation=None):
    with tf.name_scope(name):
        n_inputs = int(X.shape[1]) #fearures number (no. inputs)
        stdev = 2 / np.sqrt(n_inputs)
        init = tf.truncated_normal((n_inputs, n_neurons), stddev=stdev) #intil weight values
        W = tf.Variable(init, name="weights")
        b = tf.Variable(tf.zeros([n_neurons]), name='bias')
        z = tf.matmul(X, W) + b  
        if activation=="relu":
            return tf.nn.relu(z) #linear regression  (activation)
        else:
            return z      
```

```python
#creating hidden layers
n_hidden1 = 300
n_hidden2 = 100
n_outputs = 10
with tf.name_scope("dnn"):
    hidden1 = neuron_layer(X, n_hidden1, "hidden1", activation="relu")
    hidden2 = neuron_layer(hidden1, n_hidden2, "hidden2", activation="relu")
    logits = neuron_layer(hidden2, n_outputs, "outputs")
```

```python
# softmax cost function: cross entropy

with tf.name_scope("loss"):
    xentropy = tf.nn.sparse_softmax_cross_entropy_with_logits(labels=y, logits=logits)
    loss = tf.reduce_mean(xentropy, name="loss")
```

```python
#gradient descent optimization
with tf.name_scope("train"):
    optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
    training_op = optimizer.minimize(loss)
```

```python
#the accuracy metric
with tf.name_scope("evaluation"):
    correct = tf.nn.in_top_k(logits, y,1)
    accuracy = tf.reduce_mean(tf.cast(correct,tf.float32))
```

```python
# create a node to intiaize all variables
init = tf.global_variables_initializer()
saver = tf.train.Saver() #save the trained model to disk
```

```python
# exection phase

#load data
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("/tmp/data/")
```

```python
n_epoches = 400
batch_size = 50
```

```python
# model trining
with tf.Session() as sess:
    init.run()
    for epoch in range(n_epoches):
        for iteration in range(mnist.train.num_examples // batch_size):
            X_batch, y_batch = mnist.train.next_batch(batch_size)
            sess.run(training_op, feed_dict={X: X_batch, y: y_batch})
        acc_train = accuracy.eval(feed_dict={X:X_batch, y:y_batch})
        acc_test = accuracy.eval(feed_dict={X:mnist.test.images, y:mnist.test.labels})
        print (epoch, "train_accuracy:", acc_train, "test_accuracy:", acc_test)
    save_path = saver.save(sess, "./my_model_final.ckpt")    
```
